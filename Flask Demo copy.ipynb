```python
from flask import Flask, request, render_template
from pymongo import MongoClient
```

```python
app = Flask(__name__)
```

```python
%time
client = MongoClient('localhost',27017)
db = client.final5400
collection_stock = db.stock_data
collection_news = db.news_data
collection_nlp = db.news_NLP
```

```python
# test connection between notebook and MongoDB 
test_results = collection_stock.find({'date': '2022-02-11'})
list(test_results)
```

```python
# test connection between notebook and MongoDB 
news_results = collection_news.find({'ticker': 'AAPL', 'date': '2022-02-11'})
list(news_results)
```

```python
# test connection between notebook and MongoDB 
nlp_results = collection_nlp.find({'date': '2023-12-11', 'ticker': 'AAPL'})
list(nlp_results)
```

```python
@app.route('/')
def index():
    return render_template('index.html')
```

```python
@app.route('/search', methods=['POST'])
def search():
    ticker = request.form['ticker']
    date = request.form['date']
    query = {'ticker': ticker, 'date': date}
    results = collection_stock.find(query)
    results2 = collection_news.find(query)
    results3 = collection_nlp.find(query)
    return render_template('results.html', results=results, results2 = results2, results3 = results3)
```

```python
if __name__ == '__main__':
    app.run(port = 5001)
```
